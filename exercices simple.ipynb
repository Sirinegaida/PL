```python
def fizzbuzz(n=100):
    """
    Print FizzBuzz sequence from 1 to n.
    
    Args:
        n: The upper limit (default is 100)
    """
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)

# Run the function
if __name__ == "__main__":
    fizzbuzz(100)
```

```python
# ========== 2. PALINDROME ==========
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]


# ========== 3. ANAGRAMME ==========
def are_anagrams(s1, s2):
    return sorted(s1.lower()) == sorted(s2.lower())


# ========== 4. FIBONACCI ==========
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b


# ========== 5. NOMBRE PREMIER ==========
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


# ========== 6. FACTORIELLE ==========
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)


# ========== 7. TRI À BULLES ==========
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr


# ========== 8. RECHERCHE BINAIRE ==========
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1


# ========== 9. INVERSION DE CHAÎNE ==========
def reverse_string(s):
    return s[::-1]


# ========== 10. SOMME DE DEUX NOMBRES (Two Sum) ==========
def two_sum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []


# ========== 11. MAXIMUM DANS UN TABLEAU ==========
def find_max(arr):
    return max(arr)


# ========== 12. COMPTER LES VOYELLES ==========
def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for char in s if char in vowels)


# ========== 13. PGCD (Plus Grand Commun Diviseur) ==========
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a


# ========== 14. DUPLIQUER UN TABLEAU ==========
def remove_duplicates(arr):
    return list(set(arr))


# ========== 15. CALCUL DE MOYENNE ==========
def average(numbers):
    return sum(numbers) / len(numbers) if numbers else 0

```
